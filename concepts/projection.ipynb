```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(313)
```

### Visualization

```python

def dot_product(a, b):
    """Calculate the dot product of two vectors."""
    return sum(a_i * b_i for a_i, b_i in zip(a, b))

def magnitude(vector):
    """Calculate the magnitude (length) of a vector."""
    return np.sqrt(sum(component**2 for component in vector))

def scalar_projection(a, b):
    """
    Calculate the scalar projection of vector a onto vector b.
    
    Args:
        a: The vector being projected
        b: The vector being projected onto
        
    Returns:
        A scalar value representing how far a extends in the direction of b
    """
    return dot_product(a, b) / magnitude(b)

def vector_projection(a, b):
    """
    Calculate the vector projection of vector a onto vector b.
    
    Args:
        a: The vector being projected
        b: The vector being projected onto
        
    Returns:
        A vector in the direction of b with magnitude equal to the scalar projection
    """
    # Calculate the scalar projection first
    scalar_proj = scalar_projection(a, b)
    
    # Convert to a vector by scaling the unit vector of b
    b_unit = b / magnitude(b)
    return scalar_proj * b_unit

# Example usage
def main():
    # Define two vectors
    vector_a = np.array([3, 4])  # Vector A
    vector_b = np.array([5, 0])  # Vector B (along x-axis)
    
    # Calculate projections
    scalar_proj = scalar_projection(vector_a, vector_b)
    vector_proj = vector_projection(vector_a, vector_b)
    
    print(f"Vector A: {vector_a}")
    print(f"Vector B: {vector_b}")
    print(f"Scalar projection of A onto B: {scalar_proj}")
    print(f"Vector projection of A onto B: {vector_proj}")
    
    # Visualize the vectors and projection
    plt.figure(figsize=(10, 6))
    
    # Plot vector A
    plt.arrow(0, 0, vector_a[0], vector_a[1], head_width=0.2, head_length=0.3, fc='blue', ec='blue', label='Vector A')
    
    # Plot vector B
    plt.arrow(0, 0, vector_b[0], vector_b[1], head_width=0.2, head_length=0.3, fc='red', ec='red', label='Vector B')
    
    # Plot the vector projection
    plt.arrow(0, 0, vector_proj[0], vector_proj[1], head_width=0.2, head_length=0.3, fc='green', ec='green', label='Vector Projection')
    
    # Add a dashed line from the tip of A to the vector projection
    plt.plot([vector_a[0], vector_proj[0]], [vector_a[1], vector_proj[1]], 'k--')
    
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    plt.grid(True, alpha=0.3)
    plt.xlim(-1, 6)
    plt.ylim(-1, 5)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Vector Projection Visualization')
    plt.legend()
    plt.axis('equal')
    plt.show()

if __name__ == '__main__':
    main()
```
