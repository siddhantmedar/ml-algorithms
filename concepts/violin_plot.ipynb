```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
data = [np.random.normal(0, 1, 100),  # First distribution
        np.random.normal(3, 1.5, 100),  # Second distribution
        np.random.normal(-2, 0.5, 100)]  # Third distribution
```

```python
# Create violin plot
plt.figure(figsize=(8, 5))
sns.violinplot(data=data)
plt.xticks([0, 1, 2], ["Group A", "Group B", "Group C"])
plt.title("Violin Plot Example")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=data, inner="quartile")
plt.xticks([0, 1, 2], ["Group A", "Group B", "Group C"])
plt.title("Customized Violin Plot")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=data, ax=axes[0])
axes[0].set_title("Box Plot")

sns.violinplot(data=data, ax=axes[1])
axes[1].set_title("Violin Plot")
plt.show()
```
