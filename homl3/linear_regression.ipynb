```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data_url = "https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"

df = pd.read_csv(data_url)
```

```python
df.shape
```

```python
df.columns
```

```python
df.plot(kind='scatter', x='GDP per capita (USD)', y='Life satisfaction')
```

```python
X = df[['GDP per capita (USD)']].values
y = df[['Life satisfaction']].values

print(f"The shape of X is {X.shape} and the shape of y is {y.shape}")
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[3.13]]) # predict for a random number
```

```python
model.intercept_, model.coef_
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life satisfaction')
plt.title('Linear Regression Fit')
plt.show()
```
