```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
n_samples = 1000
n_features = 1
```

```python
X = np.random.rand(n_samples,n_features)
y = 3 + 2 * X + np.random.randn(n_samples,n_features)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y) # fit the model
```

```python
lr.intercept_[0], lr.coef_[0][0]
```

```python
# Plot the data points
plt.scatter(X, y, label='Data points')

# Plot the regression line
x_values = np.linspace(0,1,n_samples)
y_values = lr.intercept_[0] + lr.coef_[0][0] * x_values


plt.plot(x_values, y_values, color='r', label='Regression line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.show()
```
