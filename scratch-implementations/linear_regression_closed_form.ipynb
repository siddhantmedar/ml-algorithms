```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import add_dummy_feature
```

```python
np.random.seed(313)
```

```python
n_samples = 1000
n_features = 1
```

```python
X = np.random.rand(n_samples,n_features)
y = 3 + 2 * X + np.random.randn(n_samples,n_features)
```

```python
X = add_dummy_feature(X) # add a column of ones to X (bias term)
X.shape
```

```python
# closed form solution
y_hat = np.linalg.inv(X.T @ X).dot(X.T).dot(y)
y_hat
```

```python
# Plot the data points
plt.scatter(X[:, 1], y, label='Data points')

# Plot the regression line
x_values = np.linspace(0,1,n_samples)
y_values = y_hat[0] + y_hat[1] * x_values
plt.plot(x_values, y_values, color='r', label='Regression line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.show()
```
