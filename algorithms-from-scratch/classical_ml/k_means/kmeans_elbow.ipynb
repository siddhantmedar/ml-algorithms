```python
from kmeans import KMeans
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
```

```python
k_values = range(1, 50)
k_values
```

```python
data, _ = make_blobs(n_samples=1000, centers=4, cluster_std=0.6, random_state=0)
```

```python
inertias = []

for k in k_values:
    model = KMeans(k)  # model before KMeans++ init
    model.fit(data)
    inertias.append(model.inertia_)
```

```python
plt.plot(k_values, inertias)
plt.xlabel("k")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```
